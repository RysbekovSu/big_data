# №2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры 

```python
import matplotlib.pyplot as plt
import numpy as np

a = 1
b = -2
c = 1

x = np.linspace(-10, 10, 400)

y = a * x**2 + b * x + c

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = ax^2 + bx + c', color='r')

plt.title('График квадратичной функции', fontsize=16)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)

# Добавляем сетку и легенду
plt.grid(True)
plt.legend()

# Отображаем график
plt.show()
```

# №3. Построение графика произвольной гармонической функции

```python
A = 1     # Амплитуда
B = 2     # Частота
C = 0     # Фаза

x = np.linspace(0, 2 * np.pi, 400)

y = A * np.sin(B * x + C)/2

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = A*sin(Bx + C)', color='g')

plt.title('Гармоническая функция', fontsize=16)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
```

# №4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
a = 1
b = -2
c = 1
x_quad = np.linspace(-10, 10, 400)
y_quad = a * x_quad**2 + b * x_quad + c

A = 1
B = 2
C = 0
x_harmonic = np.linspace(0, 2 * np.pi, 400)
y_harmonic = A * np.sin(B * x_harmonic + C)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # 1 строка, 2 столбца, 1-я позиция
plt.plot(x_quad, y_quad, label='y = ax^2 + bx + c', color='b')
plt.title('Квадратичная функция')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)

plt.subplot(1, 2, 2)  # 1 строка, 2 столбца, 2-я позиция
plt.plot(x_harmonic, y_harmonic, label='y = A*sin(Bx + C)', color='g')
plt.title('Гармоническая функция')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)


plt.tight_layout()
plt.show()
```

# №5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
import matplotlib.pyplot as plt
import numpy as np

a = 1
b = -2
c = 1
x_quad = np.linspace(-10, 10, 400)
y_quad = a * x_quad**2 + b * x_quad + c

A = 1
B = 2
C = 0
x_harmonic = np.linspace(0, 2 * np.pi, 400)
y_harmonic = A * np.sin(B * x_harmonic + C)

fig, ax1 = plt.figure(figsize=(8, 6)), plt.gca()

ax1.plot(x_quad, y_quad, label='Квадратичная функция', color='b')
ax1.set_xlabel('Ось X')
ax1.set_ylabel('Квадратичная функция', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(x_harmonic, y_harmonic, label='Гармоническая функция', color='g')
ax2.set_ylabel('Гармоническая функция', color='g')
ax2.tick_params(axis='y', labelcolor='g')

ax1.set_ylim([-10, 50])   
ax2.set_ylim([-2, 2])     

plt.title('Квадратичная и гармоническая функции с разными масштабами')
ax1.grid(True)

plt.show()
```

# №6. Построить график функции, заданной параметрически, 
$$
x(t) = \cos(3t), \quad y(t) = \sin(2t), \quad t \in [0, 4\pi]
$$


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi, 500)

x = np.cos(3 * t)
y = np.sin(2 * t)

plt.figure(figsize=(6, 6))
plt.plot(x, y, label='x(t) = cos(3t), y(t) = sin(2t)', color='b')

plt.title('Параметрическая функция')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()

plt.show()
```

# №7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице

```python
import matplotlib.pyplot as plt
import numpy as np

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
minimum_wage = [
    4341.15, 4599.21, 4981.51, 5182.99, 
    4794.34, 4900.79, 4792.54, 4806.32, 
    5358.53, 6268.31, 7178.32
]

plt.figure(figsize=(10, 6))
plt.bar(years, minimum_wage, color='skyblue')

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)', fontsize=14)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Прожиточный минимум (сом)', fontsize=12)
plt.xticks(years) 

for i, value in enumerate(minimum_wage):
    plt.text(years[i], value + 50, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# №8.	Построить круговой график по данным п. 7

```python
import matplotlib.pyplot as plt

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
minimum_wage = [
    4341.15, 4599.21, 4981.51, 5182.99, 
    4794.34, 4900.79, 4792.54, 4806.32, 
    5358.53, 6268.31, 7178.32
]

plt.figure(figsize=(8, 8))
plt.pie(minimum_wage, labels=years, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)', fontsize=14)

plt.axis('equal') 
plt.show()
```

# №9.	Построить график произвольной каркасной поверхности

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

radius = 1    # Радиус
height = 2    # Высота
resolution = 100  # Разрешение

theta = np.linspace(0, 2 * np.pi, resolution)
z = np.linspace(0, height, resolution)
Theta, Z = np.meshgrid(theta, z)

X = radius * np.cos(Theta)
Y = radius * np.sin(Theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='blue', alpha=0.7)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Wireframe Cylinder')

ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([0, 2])

plt.show()
```

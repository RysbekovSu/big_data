```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
import mglearn
import mglearn.datasets
from sklearn.datasets import load_diabetes, load_linnerud
```

# 2.	 Подключите датасеты моделей регрессии:

```python
diabetes = load_diabetes(as_frame=True)
linnerud = load_linnerud(as_frame=True)
```

# 3	 Изучение моделей:

# 3.1 данных диабета

```python
print("Ключи breast_cancer_dataset: \n{}".format(diabetes.keys()))
print(diabetes['DESCR'][:1000] + "\n...")
print("Форма массива для Diabets: \n{}".format(diabetes.data.shape))
print("Type of the array : {}".format(type(diabetes['data'])))
print("First 10 lines of the array: \n{}".format(diabetes['data'][:10]))

print("Type of the array : {}".format(type(diabetes['target'])))
print("Shape of the array : {}".format(diabetes['target'].shape))
print("Answers : {}".format(diabetes['target'][:100]))
import pandas as pd

X = diabetes.data
y = diabetes.target
feature_names = diabetes.feature_names

data = pd.DataFrame(X, columns=feature_names)
data['target'] = y

plt.figure(figsize=(15, 10))
for i, feature in enumerate(feature_names):
    plt.subplot(2, 5, i + 1)
    plt.scatter(data[feature], data['target'], alpha=0.6)
    plt.xlabel(feature)
    plt.ylabel("Target")
    plt.title(f"{feature} c Target")

plt.tight_layout()
plt.show()
```

# 3.2 Данные физических упражнений

```python
print("Ключи breast_cancer_dataset: \n{}".format(linnerud.keys()))
print(linnerud['DESCR'][:1000] + "\n...")
print("Форма массива для Diabets: \n{}".format(linnerud.data.shape))
print("Type of the array : {}".format(type(linnerud['data'])))
print("First 10 lines of the array: \n{}".format(linnerud['data'][:10]))

print("Type of the array : {}".format(type(linnerud['target'])))
print("Shape of the array : {}".format(linnerud['target'].shape))
print("Answers : {}".format(linnerud['target'][:100]))
X = linnerud.data
y = linnerud.target
feature_names = linnerud.feature_names  
target_names = linnerud.target_names    

data_X = pd.DataFrame(X, columns=feature_names)
data_y = pd.DataFrame(y, columns=target_names)

plt.figure(figsize=(15, 10))
for i, exercise in enumerate(target_names):
    for j, feature in enumerate(feature_names):
        plt.subplot(3, 3, i * 3 + j + 1)
        plt.scatter(data_X[feature], data_y[exercise], alpha=0.6)
        plt.xlabel(feature)
        plt.ylabel(exercise)
        plt.title(f"{exercise} vs {feature}")

plt.tight_layout()
plt.show()
```

# 4.Модели регрессии (KNeighborsRegressor, LinearRegression, Ridge, Lasso, LogisticRegression) 

# 4.1 KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

X = diabetes.data  
y = diabetes.target  
X_train, X_test, y_train, y_test = train_test_split(X, y,  random_state=42)
model1 = KNeighborsRegressor(n_neighbors=1)
model3 = KNeighborsRegressor(n_neighbors=3)
model10 = KNeighborsRegressor(n_neighbors=10)
model100 = KNeighborsRegressor(n_neighbors=100)

model1.fit(X_train, y_train)
model3.fit(X_train, y_train)
model10.fit(X_train, y_train)
model100.fit(X_train, y_train)
print("Точность модели c 1 соседом соседями: {:.2}".format(model1.score(X_test,y_test)))
print("Точность модели c 3мя соседями: {:.2}".format(model3.score(X_test,y_test)))
print("Точность модели с 10 соседями: {:.2}".format(model10.score(X_test,y_test)))
print("Точность модели с 100 соседями: {:.2}".format(model100.score(X_test,y_test)))
y_pred = model10.predict(X_test)
print("\nПервые 5 предсказанных значений:\n", y_pred[:5])
print("Первые 5 реальных значений:\n", y_test[:5])
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

X2 = linnerud.data  
y2 = linnerud.target  
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2,  random_state=42)
lmodel1 = KNeighborsRegressor(n_neighbors=1)
lmodel3 = KNeighborsRegressor(n_neighbors=3)
lmodel10 = KNeighborsRegressor(n_neighbors=10)

lmodel1.fit(X_train2, y_train2)
lmodel3.fit(X_train2, y_train2)
lmodel10.fit(X_train2, y_train2)
print("Точность модели c 3мя соседями: {:.2}".format(lmodel3.score(X_test2,y_test2)))
print("Точность модели с 10 соседями: {:.2}".format(lmodel10.score(X_test2,y_test2)))
print("Точность модели c 1 соседом соседями: {:.2}".format(lmodel1.score(X_test2,y_test2)))

y_pred = lmodel10.predict(X_test2)
print("\nПервые 5 предсказанных значений:\n", y_pred[:5])
print("Первые 5 реальных значений:\n", y_test2[:5])
```

# 4.2 LinearRegression

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression().fit(X_train, y_train)
print("Точность модели линейной регрессии на обучающих: {:.2f}".format(linear.score(X_train, y_train)))
print("Точность модели линейной регрессии на тестовых данных: {:.2f}".format(linear.score(X_test, y_test)))
print("linear coef_: {}".format(linear.coef_))
print("linear intercept_: {:.2}".format(linear.intercept_))

y_pred = linear.predict(X_test)
print("\nПервые 5 предсказанных значений:\n", y_pred[:5])
print("Первые 5 реальных значений:\n", y_test[:5])
```
